# Data Cleaning

Raw business data often requires preparation before it can be used for analysis. The objective of this phase is to improve data quality while preserving the integrity of the original information.

The cleaning process includes:

- Creating a working copy of the dataset
- Handling missing values
- Removing duplicate records
- Standardizing text fields
- Correcting data types
- Validating numerical values
- Reviewing the cleaned dataset

## Create a Working Copy

To preserve the original dataset, a copy is created for all cleaning and transformation tasks. The original dataset remains unchanged throughout the project.

```python
# Create Working Copy
df_clean = df.copy()
print("Working copy created successfully.")
```

## Handle Missing Values

Missing values can affect analysis and reporting. The following steps fill missing values using appropriate methods based on the type of data in each column.

```python
# Handle Missing Values

# Fill missing age values with the median age
df_clean["Age"] = df_clean["Age"].fillna(df_clean["Age"].median())

# Fill missing discount values with 0
df_clean["Discount_%"] = df_clean["Discount_%"].fillna(0)

# Fill missing customer ratings with the median rating
df_clean["Customer_Rating"] = df_clean["Customer_Rating"].fillna(
    df_clean["Customer_Rating"].median()
)

# Fill missing payment methods with the most frequent value
df_clean["Payment_Method"] = df_clean["Payment_Method"].fillna(
    df_clean["Payment_Method"].mode()[0]
)

# Fill missing loyalty status with the most frequent value
df_clean["Loyalty_Status"] = df_clean["Loyalty_Status"].fillna(
    df_clean["Loyalty_Status"].mode()[0]
)

print("Missing values handled successfully.")
```

## Remove Duplicate Records

Duplicate transactions can lead to inaccurate business metrics. Removing duplicate records ensures each transaction is counted only once.

```python
# Remove Duplicate Records

rows_before = len(df_clean)

# Remove duplicate rows
df_clean.drop_duplicates(inplace=True, ignore_index=True)

rows_after = len(df_clean)

print(f"Rows Before          : {rows_before}")
print(f"Rows After           : {rows_after}")
print(f"Duplicates Removed   : {rows_before - rows_after}")
```

## Standardize Text Values

Text fields are standardized to improve consistency across the dataset. This includes removing extra spaces and applying consistent capitalization.

```python
# Clean Text Columns

text_columns = df_clean.select_dtypes(include="object").columns

for column in text_columns:
    df_clean[column] = (
        df_clean[column]
        .astype(str)
        .str.strip()
        .str.title()
    )

print("Text values standardized successfully.")
```

## Convert Data Types

Columns are converted to appropriate data types to ensure accurate calculations and efficient analysis.

```python
# Convert Data Types

df_clean["Order_Date"] = pd.to_datetime(
    df_clean["Order_Date"],
    errors="coerce"
)

print(df_clean.dtypes)
```

## Validate Numeric Values

Business rules are applied to identify records containing unrealistic numerical values. These records are retained for review while values outside acceptable ranges are replaced with missing values for later treatment.

```python
# Validate Numeric Values

df_clean.loc[df_clean["Quantity"] <= 0, "Quantity"] = np.nan
df_clean.loc[df_clean["Unit_Price_SAR"] <= 0, "Unit_Price_SAR"] = np.nan
df_clean.loc[~df_clean["Age"].between(18, 75), "Age"] = np.nan
df_clean.loc[~df_clean["Customer_Rating"].between(1, 5), "Customer_Rating"] = np.nan

print("Numeric validation completed.")
```

## Fill Values After Validation

Some values were replaced during validation because they fell outside acceptable business ranges. These missing values are filled using suitable statistical measures.

```python
# Fill Values After Validation

df_clean["Quantity"] = df_clean["Quantity"].fillna(df_clean["Quantity"].median())
df_clean["Unit_Price_SAR"] = df_clean["Unit_Price_SAR"].fillna(df_clean["Unit_Price_SAR"].median())
df_clean["Age"] = df_clean["Age"].fillna(df_clean["Age"].median())
df_clean["Customer_Rating"] = df_clean["Customer_Rating"].fillna(df_clean["Customer_Rating"].median())

print("Validated values updated successfully.")
```

## Verify the Cleaned Dataset

After completing the cleaning process, the dataset is reviewed to confirm that the identified issues have been addressed.

```python
# Verify Cleaned Dataset

print("Missing Values:\n")
print(df_clean.isnull().sum())

print("\nDuplicate Records:")
print(df_clean.duplicated().sum())

print("\nDataset Shape:")
print(df_clean.shape)
```

## Data Cleaning Summary

The dataset has been prepared for analysis through a structured cleaning process. Key data quality issues identified during the assessment phase have been addressed, resulting in a more consistent and reliable dataset.

# Feature Engineering

Feature engineering is the process of creating new variables from existing data to improve analysis and support business decision-making.

The new features created in this phase will help analyze sales trends, customer behavior, profitability, and product performance from different perspectives.

## Create Date-Based Features

Date-related features make it possible to analyze sales across different time periods, such as years, months, quarters, weekdays, and weekends.

```python
# Create Date Features

df_clean["Year"] = df_clean["Order_Date"].dt.year.astype("Int64")
df_clean["Month"] = df_clean["Order_Date"].dt.month.astype("Int64")
df_clean["Month_Name"] = df_clean["Order_Date"].dt.month_name()
df_clean["Quarter"] = df_clean["Order_Date"].dt.quarter.astype("Int64")
df_clean["Day"] = df_clean["Order_Date"].dt.day.astype("Int64")
df_clean["Weekday"] = df_clean["Order_Date"].dt.day_name()

print("Date features created successfully.")
```

## Identify Weekend Transactions

Retail purchasing patterns often differ between weekdays and weekends. A new feature is created to distinguish these transactions.

```python
# Create Weekend Indicator
df_clean["Is_Weekend"] = df_clean["Weekday"].isin(
    ["Friday", "Saturday"]
)

print("Weekend indicator created successfully.")
```

## Calculate Profit Margin

Profit margin measures how much profit is earned from each Saudi Riyal of sales and is a useful indicator of business performance.

```python
# Calculate Profit Margin
df_clean["Profit_Margin_%"] = (
    df_clean["Profit_SAR"] /
    df_clean["Net_Sales_SAR"]
) * 100

df_clean["Profit_Margin_%"] = df_clean["Profit_Margin_%"].round(2)

print("Profit margin calculated successfully.")
```

## Categorize Customers by Age

Grouping customers into age categories simplifies customer segmentation and helps identify purchasing patterns across different age groups.

```python
# Create Age Groups
age_bins = [17, 25, 35, 45, 55, 65, 100]
age_labels = [
    "18-25",
    "26-35",
    "36-45",
    "46-55",
    "56-65",
    "65+"
]

df_clean["Age_Group"] = pd.cut(
    df_clean["Age"],
    bins=age_bins,
    labels=age_labels
)

print("Age groups created successfully.")
```

## Classify Order Value

Order value categories provide a simple way to compare purchasing behavior across low, medium, and high-value transactions.

```python
# Create Order Value Categories
order_bins = [0, 500, 1500, float("inf")]
order_labels = [
    "Low",
    "Medium",
    "High"
]

df_clean["Order_Value"] = pd.cut(
    df_clean["Net_Sales_SAR"],
    bins=order_bins,
    labels=order_labels
)

print("Order value categories created successfully.")
```

## Review Newly Created Features

The final step is to review the newly created variables to ensure they have been added correctly before moving to exploratory data analysis.

```python
# Display Newly Created Features

new_features = [
    "Year",
    "Month",
    "Month_Name",
    "Quarter",
    "Weekday",
    "Is_Weekend",
    "Profit_Margin_%",
    "Age_Group",
    "Order_Value"
]

df_clean[new_features].head(10)
```
